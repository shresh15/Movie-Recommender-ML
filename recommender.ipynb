```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('./movies_dataset/tmdb_5000_movies.csv')
credits=pd.read_csv('./movies_dataset/tmdb_5000_credits.csv')
```

```python
movies = pd.merge(movies, credits, on='title')
```

```python
#genres id keywords title overview cast crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]  
```

To get the tags, we shall merge overview genres keywords cast crew

```python
#detecting nulls
movies.isnull().sum()
```

```python
#dropping the nulls
movies.dropna(inplace=True)
```

```python
#checking duplicates
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
#We have : '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#we want ['Action','Adventure','Fantasy','SciFi']
import ast
def convert(obj): # here object is a string
    L=[]
    for i in ast.literal_eval( obj): # converting a string to list
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

For cast: Get the first 3 actors from the dictionary
character is the movie char. name
GET THE NAME


```python
import ast
def convert3(obj): # here object is a string
    L=[]
    counter=0
    for i in ast.literal_eval( obj):
         # converting a string to list
        if (counter!=3):
            L.append(i['name'])
            counter=counter+1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

For crew, only fetch director name 
job:director


```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L

            
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#converting the overview from string to a list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

"Sam" "Mendes" And "Sam" "Worthington" The model might get confused: as Sam and Sam
so-> SamMendes works fine as 1 entity


```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```
